```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests as rq

url = "https://www.tanzapages.com/location/Dar_Es_Salaam"
all_dataframes = []

while True:
    req = rq.get(url, allow_redirects=False)
    req_content = req.content
    soup = bs(req_content, "html.parser")

    docu = soup.find(id="listings").find_all(class_="company")

    for i in docu:
        if i.a and i.a.get('href') is not None:
            company_url = f"https://www.tanzapages.com{i.a['href']}"

            # Manually handle redirects
            try:
                req_company = rq.get(company_url, allow_redirects=False)
                req_company_content = req_company.content
                
                # Check for the presence of a Location header
                if 'Location' in req_company.headers:
                    new_url = req_company.headers['Location']
                    req_company = rq.get(new_url)
                    req_company_content = req_company.content
            except rq.exceptions.TooManyRedirects:
                print(f"Too many redirects for '{company_url}'. Skipping.")
                continue

            soup_company = bs(req_company_content, "html.parser")

            # Extract company name
            company_name = soup_company.find("h1").text.strip()

            # Extract company details
            details_container = soup_company.find(class_="cmp_details_in")

            if details_container:
                details = details_container.find_all(class_="info")

                # Create a dictionary to store the details
                details_dict = {"Company Name": company_name}

                for detail in details:
                    label_element = detail.find(class_="label")
                    text_element = detail.find(class_="text")

                    if label_element and text_element:
                        label = label_element.text.strip()
                        text = text_element.text.strip()

                        # Add the detail to the dictionary
                        details_dict[label] = text

                # Create a DataFrame from the dictionary
                df = pd.DataFrame([details_dict])
                all_dataframes.append(df)

                print(f"Details for '{company_name}' added to the list.")
            else:
                print(f"No company details found on the page for '{company_name}'.")

    # Find the next page link
    pages_container = soup.find(class_="pages_container")
    next_link = pages_container.find("a", rel="next")

    if next_link:
        url = f"https://www.tanzapages.com{next_link['href']}"
    else:
        print("No more pages found. Exiting the loop.")
        break

# Concatenate all DataFrames into a single DataFrame
final_df = pd.concat(all_dataframes, ignore_index=True)

# Export the final DataFrame to CSV
final_df.to_csv('AllCompanies.csv', index=False)

# Display the final DataFrame
print(final_df)
```
